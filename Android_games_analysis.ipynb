```python
#Join columns API_Ratings and Average Ratings. If NAN value in API_ratings then Average Ratings value will be kept. API_Ratings value is more 
#accurate than Average Ratings, which is why this merge is being done.The 'Is_English' column will be excluded since it has no valuable information. 
conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS merged_dataset4;")

cursor.execute("""
CREATE TABLE merged_dataset4 AS 
SELECT 
    COALESCE(a.Game, w.Game, k.Game) AS Game_Title,
    COALESCE(a.Category, w.Category, k.Category) AS Category,
    COALESCE(a."Total Ratings", w."Total Ratings", k."Total Ratings") AS "Total Ratings",
    -- API_Ratings takes precedence over Average Rating
    COALESCE(a."API_Ratings", a."Average Rating", w."Average Rating", k."Average Rating") AS "Final Rating",
    COALESCE(a."5 Star Rating", w."5 Star Rating", k."5 Star Rating") AS "5 Star Rating",
    COALESCE(a."4 Star Rating", w."4 Star Rating", k."4 Star Rating") AS "4 Star Rating",
    COALESCE(a."3 Star Rating", w."3 Star Rating", k."3 Star Rating") AS "3 Star Rating",
    COALESCE(a."2 Star Rating", w."2 Star Rating", k."2 Star Rating") AS "2 Star Rating",
    COALESCE(a."1 Star Rating", w."1 Star Rating", k."1 Star Rating") AS "1 Star Rating",
    COALESCE(a."Final_Title", w."Final_Title", k."Final_Title") AS "Final_Title",
    COALESCE(a."Release", w."Release") AS "Release",
    COALESCE(a."Revenue", w."Revenue") AS "Revenue",
    COALESCE(a."Company", w."Company") AS "Company",
    COALESCE(a."Total Downloads", 'Not Available') AS "Total Downloads", 
    COALESCE(a."Low Rating & High Installs", 'Not Available') AS "Low Rating & High Installs",
    COALESCE(a."Content Rating", 'Not Available') AS "Content Rating"
FROM API_table a
LEFT JOIN web_scrape_table w ON a.Game = w.Game
LEFT JOIN kaggle_data k ON a.Game = k.Game;
""")

conn.commit()
conn.close()

print("Merged_dataset4 made")
```

```python
#Pie chart of distribution of Content Ratings:
import matplotlib.pyplot as plt
import seaborn as sns
conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

#get content rating counts 
cursor.execute("""
    SELECT "Content Rating", COUNT(*) 
    FROM merged_dataset4 
    WHERE "Content Rating" NOT IN ('Not Available', 'Not found')
    GROUP BY "Content Rating";
""")
data = cursor.fetchall()
conn.close()

ratings, counts = zip(*data)

plt.figure(figsize=(8,8))
plt.pie(counts, labels=ratings, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
#update True Ratings column
conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS merged_dataset4;")

cursor.execute("""
CREATE TABLE merged_dataset4 AS 
SELECT 
    COALESCE(a.Game, w.Game, k.Game) AS Game_Title,
    COALESCE(a.Category, w.Category, k.Category) AS Category,
    COALESCE(a."Total Ratings", w."Total Ratings", k."Total Ratings") AS "Total Ratings",

    -- Corrected True Rating logic: API_Ratings takes priority, but falls back to Average Rating if 'Not found'
    CASE 
        WHEN a."API_Ratings" = 'Not found' OR a."API_Ratings" IS NULL 
        THEN COALESCE(a."Average Rating", w."Average Rating", k."Average Rating") 
        ELSE a."API_Ratings"
    END AS "True Rating",

    COALESCE(a."5 Star Rating", w."5 Star Rating", k."5 Star Rating") AS "5 Star Rating",
    COALESCE(a."4 Star Rating", w."4 Star Rating", k."4 Star Rating") AS "4 Star Rating",
    COALESCE(a."3 Star Rating", w."3 Star Rating", k."3 Star Rating") AS "3 Star Rating",
    COALESCE(a."2 Star Rating", w."2 Star Rating", k."2 Star Rating") AS "2 Star Rating",
    COALESCE(a."1 Star Rating", w."1 Star Rating", k."1 Star Rating") AS "1 Star Rating",
    COALESCE(a."Final_Title", w."Final_Title", k."Final_Title") AS "Final_Title",
    COALESCE(a."Release", w."Release") AS "Release",
    COALESCE(a."Revenue", w."Revenue") AS "Revenue",
    COALESCE(a."Company", w."Company") AS "Company",
    COALESCE(a."Total Downloads", 'Not Available') AS "Total Downloads", 
    COALESCE(a."Low Rating & High Installs", 'Not Available') AS "Low Rating & High Installs",
    COALESCE(a."Content Rating", 'Not Available') AS "Content Rating"
FROM API_table a
LEFT JOIN web_scrape_table w ON a.Game = w.Game
LEFT JOIN kaggle_data k ON a.Game = k.Game;
""")


conn.commit()
conn.close()

print("`merged_dataset4` updated with corrected 'True Rating' in Milestone_five.db")
```

```python
#Box plot of game ratings for each gaming category: 
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

#get ratings by category
cursor.execute("""
    SELECT "Category", CAST("True Rating" AS FLOAT) 
    FROM merged_dataset4
    WHERE "True Rating" IS NOT NULL
    ORDER BY "Category";
""")
data = cursor.fetchall()
conn.close()

if not data:
    print("No valid rating data available!")
else:
    categories = []
    ratings = []
    
    for category, rating in data:
        categories.append(category)
        ratings.append(rating)

#plot boxplot
    plt.figure(figsize=(12,6))
    sns.boxplot(x=categories, y=ratings)
    plt.xticks(rotation=45, ha="right")  # Rotate category labels for readability
    plt.xlabel("Game Category")
    plt.ylabel("True Rating")
    plt.title("Distribution of Game Ratings Across Game Categories")
    plt.show()
```

```python
#Do bigger companies get more downloads and ratings? 
import sqlite3
import matplotlib.pyplot as plt


conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

#company data
cursor.execute("""
    SELECT "Company", 
           AVG(CAST("Total Ratings" AS FLOAT)), 
           AVG(CAST("Total Downloads" AS FLOAT)), 
           COUNT(DISTINCT "Game_Title")
    FROM merged_dataset4
    WHERE "Total Downloads" != 'Not Available' AND "Total Ratings" IS NOT NULL
    GROUP BY "Company"
    HAVING COUNT(DISTINCT "Game_Title") > 1
    ORDER BY AVG(CAST("Total Downloads" AS FLOAT)) DESC
    LIMIT 20;
""")
data = cursor.fetchall()
conn.close()


companies, avg_ratings, avg_downloads, game_counts = zip(*data)

#bubble chart
plt.figure(figsize=(12,6))
plt.scatter(avg_downloads, avg_ratings, s=[g*50 for g in game_counts], alpha=0.6, color="blue", edgecolors="black")
plt.xlabel("Average Total Downloads")
plt.ylabel("Average Total Ratings")
plt.title("Do Bigger Companies Get More Downloads & Ratings?")
plt.xscale("log")  # Log scale for better visualization
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

#attatch key company 
for i, company in enumerate(companies[:10]):  # Only labeling the top 10 for clarity
    plt.annotate(company, (avg_downloads[i], avg_ratings[i]), fontsize=9, ha="right")

plt.show()
```

```python
#Do games with low ratings have a high number of installs? 
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

#Games with "Low Ratings & High Installs" 
cursor.execute("""
    SELECT CAST("Total Downloads" AS INTEGER), CAST("True Rating" AS FLOAT), "Low Rating & High Installs"
    FROM merged_dataset4
    WHERE "Total Downloads" != 'Not Available' AND "True Rating" IS NOT NULL;
""")
data = cursor.fetchall()
conn.close()

downloads, ratings, labels = zip(*data)

# Convert "Yes"/"No" to color coding
colors = ["red" if label == "Yes" else "blue" for label in labels]

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(downloads, ratings, c=colors, alpha=0.7, label="Games")
plt.xscale("log")  # Log scale to handle large download numbers
plt.xlabel("Total Downloads (Log Scale)")
plt.ylabel("True Rating")
plt.title("Do Low-Rated Games Have High Installs?")
plt.grid(True, linestyle="--", linewidth=0.5)

plt.scatter([], [], c="red", label="Low Rating & High Installs", alpha=0.7)
plt.scatter([], [], c="blue", label="Other Games", alpha=0.7)
plt.legend()

plt.show()
```

```python
#Top 10 games: 5 star vs. 1 star ratings 
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

conn = sqlite3.connect("Milestone_five.db")
cursor = conn.cursor()

#top 10 games with the most 5-star ratings
cursor.execute("""
    SELECT "Game_Title", CAST("5 Star Rating" AS INTEGER)
    FROM merged_dataset4
    ORDER BY CAST("5 Star Rating" AS INTEGER) DESC
    LIMIT 10;
""")
five_star_data = cursor.fetchall()

#top 10 games with the most 1-star ratings
cursor.execute("""
    SELECT "Game_Title", CAST("1 Star Rating" AS INTEGER)
    FROM merged_dataset4
    ORDER BY CAST("1 Star Rating" AS INTEGER) DESC
    LIMIT 10;
""")
one_star_data = cursor.fetchall()

conn.close()

five_star_games, five_star_counts = zip(*five_star_data)
one_star_games, one_star_counts = zip(*one_star_data)

#combined figure
fig, ax = plt.subplots(figsize=(12,6))

#bar width
bar_width = 0.4
y_positions = np.arange(len(five_star_games))

# Plot 5star and 1star ratings side by side
ax.barh(y_positions - bar_width/2, five_star_counts, height=bar_width, color="navy", label="5-Star Ratings")
ax.barh(y_positions + bar_width/2, one_star_counts, height=bar_width, color="red", label="1-Star Ratings")

ax.set_yticks(y_positions)
ax.set_yticklabels(five_star_games)
ax.set_xlabel("Number of Ratings")
ax.set_ylabel("Game Title")
ax.set_title("Top 10 Games with Most 5-Star vs. 1-Star Ratings")
ax.legend()

ax.invert_yaxis()

plt.show()
```
